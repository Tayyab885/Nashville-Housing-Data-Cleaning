### Features with Missing Values

```python
features_na = [feature for feature in df.columns if df[feature].isnull().sum()>=1]
features_na
```

```python
for i in features_na:
    print(f"{i} has {np.round(df[i].isnull().mean(),3)}% missing values")
```

### Exploratory Data Analysis

```python
### Checking missing values with respect to SalePrice
for i in features_na:
    data = df.copy()
    data[i] = np.where(data[i].isnull(),1,0)
    data.groupby(i)['SalePrice'].median().plot.bar()
    plt.title(i)
    plt.ylabel('SalePrice')
    plt.show()
```

#### Objective
- We can see that the missing values have great relationship with our target variable. So, we will have to fill the missing values with the appropriate/meaningful values.

### Cleaning Categorical Features

```python
df['SoldAsVacant'] = df['SoldAsVacant'].str.replace('N','No').replace('Y','Yes')
```

```python
df['OwnerName'] = df['OwnerName'].fillna('Unknown')
```

```python
df['OwnerAddress'] = df['OwnerAddress'].fillna('Unknown')
```

```python
median = df['Acreage'].median()
df['Acreage'] = df['Acreage'].fillna(median)
```

```python
mode = str(df['TaxDistrict'].mode().values[0])
df['TaxDistrict'] = df['TaxDistrict'].fillna(mode)
```

```python
df['LandValue'] = df['LandValue'].fillna(df['LandValue'].median())
```

```python
df['Bedrooms'] = df['Bedrooms'].fillna(df['Bedrooms'].median())
df['FullBath'] = df['FullBath'].fillna(df['FullBath'].median())
df['HalfBath']  = df['HalfBath'].fillna(df['HalfBath'].median())
```
